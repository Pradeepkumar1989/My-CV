```python
# importing pandas module
import pandas as pd
  
# making data frame from csv file
data = pd.read_csv("WeatherData.csv", index_col ="datetime" )
  
# dropping passed columns
data.drop([ "dew", "precip", "windspeed", "windgust", "sealevelpressure", "winddir"], axis = 1, inplace = True)
  
# display
data
```

```python
data["temperature"] = data["temp"]
data["humidity"] = data["humidity"]
labels = data.icon.unique()
labels
```

```python
processed_data = data.copy()

norain = ['partly-cloudy-day', 'clear-day''wind' ]
mightrain = ['partly-cloudy-day','rain', 'wind']
rain = ['rain']

processed_data["NoRain"] = 0
processed_data["MightRain"] = 0
processed_data["Rain"] = 0

processed_data["NoRain"].loc[data.icon.isin(norain)] = 1
processed_data["MightRain"].loc[data.icon.isin(mightrain)] = 1
processed_data["Rain"].loc[data.icon.isin(rain)] = 1
processed_data = processed_data.drop(columns="icon")

processed_data
```

```python
y = processed_data[["NoRain", "MightRain", "Rain"]].to_numpy()[7:]
rawx = processed_data.drop(columns = ["NoRain", "MightRain", "Rain"]).to_numpy()

x = []
temp = np.array(0)
for i in range(len(rawx)-7):
    temp = rawx[i:i+6].flatten()
    x.append(temp)
x = np.array(x)
x.shape, y.shape
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from keras.layers import Dense
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20, random_state = 33)
```

```python
model = tf.keras.Sequential()
model.add(Dense(14, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss = 'categorical_crossentropy',
              optimizer = tf.optimizers.Adam(),
              metrics=['accuracy'])

model.fit(xtrain, ytrain, epochs=10, validation_split=0.1)

test_results = model.evaluate(xtest, ytest, verbose=1)

model.save('model')
```

```python
# Convert the model to the TensorFlow Lite format without quantization
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the model to disk
open("model.tflite", "wb").write(tflite_model)
  
import os
basic_model_size = os.path.getsize("model.tflite")
print("Model is %d bytes" % basic_model_size)
```
